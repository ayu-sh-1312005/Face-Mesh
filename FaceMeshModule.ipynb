```python
import cv2
import mediapipe as mp
import time
def rescale_frame(frame, percent=75):
        width = int(frame.shape[1] * percent/ 100)
        height = int(frame.shape[0] * percent/ 100)
        dim = (width, height)
        return cv2.resize(frame, dim, interpolation =cv2.INTER_AREA)

class FindMeshDetector():
    def __init__(self,mode=False,max_num_faces=5,minDetectionCon=0.5,minTrackingCon=0.5):
        self.mode=mode
        self.max_num_faces=max_num_faces
        self.minDetectionCon=minDetectionCon
        self.minTrackingCon=minTrackingCon
        self.mpDraw=mp.solutions.drawing_utils
        self.drawSpec = self.mpDraw.DrawingSpec(thickness=1, circle_radius=2,color=(0,255,0))
        self.mpFaceMesh=mp.solutions.face_mesh
        self.faceMesh=self.mpFaceMesh.FaceMesh()
    def findFaceMesh(self,img,draw=True):
        imgRGB=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        self.results=self.faceMesh.process(imgRGB)
        faces=[]
        if self.results.multi_face_landmarks:
            for lm in self.results.multi_face_landmarks:
                # FACE_CONNECTIONS is renamed to FACEMESH_CONTOURS.
                ih,iw,ic=img.shape
                face=[]
                for id,ilm in enumerate(lm.landmark):
                    x,y=int(ilm.x*iw),int(ilm.y*ih)
                    print(id,x,y)
                    face.append([id,x,y])
                faces.append(face)
                if draw:
                    self.mpDraw.draw_landmarks(img,lm,self.mpFaceMesh.FACEMESH_CONTOURS,landmark_drawing_spec=self.drawSpec)
        return img,faces
    
def main():
    cap=cv2.VideoCapture(0)
    cTime=0
    pTime=0
    detector=FindMeshDetector()
    while True:
            success,img=cap.read()
            img,faces=detector.findFaceMesh(img)
            # fps
            cTime=time.time()
            fps=1/(cTime-pTime)
            pTime=cTime
            cv2.putText(img,str(int(fps)),(20,70),cv2.FONT_HERSHEY_PLAIN,3,(0,255,0),5)

            cv2.imshow("img",rescale_frame(img,100))
            if cv2.waitKey(1) & 0xFF==ord('a'):
                break
    cap.release()
    cv2.destroyAllWindows()
if __name__=='__main__':
    main()

                
        
```
